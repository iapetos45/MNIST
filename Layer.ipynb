```python
class MulLayer:
    def __init__(self):
        self.x = none
        self.y = none
    def forward(self, x, y): # 순전파
        self.x = x
        self.y = y
        out = x * y
        return out
    def backward(self, dout): #역전파
        dx = dout * self.y
        dy = dout * self.x
        return dx, dy
```

```python
class Sigmoid:
    def __init__(self):
        self.out = None
    def forward(self, x):
        out = 1 / 1(1 + np.exp(-x))
        self.out = out
        return out
    def backward(self, dout):
        dx = dout * (1.0 - self.out)* self.out
        return dx
```
